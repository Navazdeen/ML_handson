```python
grpd = Hotel_data.groupby(['type_of_meal_plan', 'booking_status'])['Booking_ID'].count().reset_index()

width=0.35
x = np.arange(grpd.shape[0]/2)
fig, ax = plt.subplots()
rect1 = ax.bar(x-width/2, grpd[grpd['booking_status']=='Canceled']['Booking_ID'], width, label='Canceled')
rect2 = ax.bar(x+width/2, grpd[grpd['booking_status']=='Not_Canceled']['Booking_ID'], width, label='Not_Canceled')

ax.set_xticks(x, grpd['type_of_meal_plan'].unique())
ax.set_ylabel("Before taking log")
ax.bar_label(rect1, padding=3)
ax.bar_label(rect2, padding=3)
ax.legend()
plt.show()
```

```python
grpd = Hotel_data.groupby(['type_of_meal_plan', 'booking_status'])['Booking_ID'].count().reset_index()
grpd['Booking_ID'] = np.log(grpd['Booking_ID'])

width=0.35
x = np.arange(grpd.shape[0]/2)
fig, ax = plt.subplots()
rect1 = ax.bar(x-width/2, grpd[grpd['booking_status']=='Canceled']['Booking_ID'], width, label='Canceled')
rect2 = ax.bar(x+width/2, grpd[grpd['booking_status']=='Not_Canceled']['Booking_ID'], width, label='Not_Canceled')

ax.set_xticks(x, grpd['type_of_meal_plan'].unique())
ax.bar_label(rect1, padding=1, fmt="%0.1f")
ax.bar_label(rect2, padding=1, fmt="%0.1f")
ax.set_ylabel("After taking log")
ax.legend()
plt.show()
```
