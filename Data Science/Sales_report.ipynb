```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./Amazon Sale Report.csv", index_col="index")
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data = data.drop("Unnamed: 22", axis=1)
```

```python
data.columns
```

```python
data.info()
```

```python
data.describe()
```

```python
data['currency'] = data['currency'].fillna("INR")
data['currency'].unique()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.info()
```

```python
data["Date"].min(), data["Date"].max()
```

```python
data["Amount"].sum()
```

```python
data['Category'].unique()
```

```python
data.groupby(by=['Category']).agg({
    "Amount":'sum',
    "Qty":'sum'
}).sort_values(by=['Amount', 'Qty'], ascending=False)
```

```python
data.groupby(by=["Size"]).agg({
    "Qty":'sum'
}).sort_values(by=['Qty'], ascending=False)
```

```python
data["Status"].unique()
```

```python
data["Courier Status"].unique()
```

```python
data.groupby(by=['Courier Status']).agg({
    "Order ID":'count'
})
```

```python
data.rename(columns={
    'Sales Channel ':'Sales Channel'
}, inplace=True)
data.columns
```

```python
data['Sales Channel'].unique()
```

```python
data.groupby(by=['Sales Channel']).agg({
    "Order ID":'count',
    "Qty":"sum",
    "Amount":'sum'
})
```

```python
data.groupby(by=['ship-country', 'ship-state', 'ship-city']).agg({
    "Order ID":'count',
    "Qty":"sum",
    "Amount":'sum'
}).to_csv("Amazon Sale Report_grouped.csv")
```

```python
data[data['Date']=='2022-03-31']
```

```python
data.groupby(by=['Date']).agg({
    "Order ID":'count',
}).sort_index()
```

```python
data['Week'] = data['Date'] - data['Date'].dt.weekday.apply(lambda x: pd.Timedelta(f"{x}day"))
```

```python
data[['Date', 'Week']].sort_values('Date').head(20)
```

```python
data.groupby(by=['Week']).aggregate({
    "Order ID":'count',
}).to_csv("excl.csv")
```
